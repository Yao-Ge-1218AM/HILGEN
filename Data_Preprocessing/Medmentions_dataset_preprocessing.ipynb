```python
with open("/content/corpus_pubtator.txt","r") as f1:
  with open("/content/dev.txt","w") as wf:
    lines = f1.readlines()
    for line in lines:
      if(line=="\n"):
        wf.write(line)
      if(line!="\n" and line[8]!="|"):
        if(line.strip("\n").split("\t")[0] in test_list):
         wf.write(line)
      elif(line!="\n" and line[8]=="|"):
        if(line.strip("\n").split("|")[0] in test_list):
          wf.write(line)
```

```python
import re
p = re.compile('\n\n',re.S)
fileContent = open("/content/dev.txt","r",encoding="utf8").read()
paraList = p.split(fileContent)

fileWriter = open("/content/dev/0.txt","a",encoding="utf8")
for paraIndex in range(len(paraList)):
  fileWriter.write(paraList[paraIndex])
  fileWriter.write("\n")
  fileWriter.close()
  fileWriter = open("/content/dev/" + str(paraIndex+1) + ".txt","a",encoding="utf8")
fileWriter.close()
print("finished")
print(paraIndex)
```

```python
import os

data_dir = "/content/dev/"
output_dir = "/content/dev-new/"
for file in os.listdir(data_dir):
  with open(data_dir + file,"r") as f:
    #print(file)
    lines = f.readlines()
    if(os.path.getsize(data_dir + file)!=0):
      file_name = lines[0].strip("\n").split("|")[0]
      for line in lines:
        if(line!="\n" and line[8]!="|"):
          with open(output_dir + file_name + ".ann","a") as wf:
            wf.write(line)
        elif(line!="\n" and line[8]=="|"):
          with open(output_dir + file_name + ".txt","a") as wf:
            wf.write(line.strip("\n").split("|")[2] + "\n")
```
